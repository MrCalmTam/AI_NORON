```python
import pandas as pd 
wl = '/content/crabs.csv'
data=pd.read_csv(wl)
data.head()

from sklearn.linear_model import Perceptron
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score # sai so , do chinh xac
from sklearn.preprocessing import StandardScaler # chuan hoa du lieu, ti le
```

```python
data = data.replace(to_replace=["B" ],value="0")
data = data.replace(to_replace=["O" ],value="1")
data = data.replace(to_replace=["M" ],value="1")
data = data.replace(to_replace=["F" ],value="0")
data
```

```python
iris=datasets.load_iris() 
X = iris.data
Y = iris.target 
```

```python
X = data.iloc[:,0:8]
Y = data.iloc[:,1]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
sc=StandardScaler()  #dat ham 
sc.fit(X_train)   
X_train_std=sc.transform(X_train) # chuyen du lieu x tran thanh std 
X_test_std=sc.transform(X_test)   # chuyen du lieu x tran thanh std 
```

```python
model = Perceptron(max_iter = 100 ,eta0 = 0.1 , random_state =0 )   # day cho no hoc , max_iter la so lan hoc, eta0 la toc do hoc tu 0 ->1.
model.fit(X_train_std,Y_train)
Y_pred=model.predict(X_test_std)
print(Y_pred)
print(Y_test)
print(' ĐỘ CHÍNH XÁC CỦA MÔ HÌNH SCALE :',accuracy_score(Y_test,Y_pred))
```
